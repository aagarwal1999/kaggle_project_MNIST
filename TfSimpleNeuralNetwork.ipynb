```python
import tensorflow as tf
import numpy as np
import pandas as pd
import scipy as sp
import scipy.io as spio
#see this paper for details on the data https://arxiv.org/pdf/1702.05373v1.pdf
k = spio.loadmat('Data/matlab/emnist-digits.mat') #change 'letters' to 'mnist', 'balanced', 'byclass', 'bymerge' or 'digits' 
```

```python
training_data = np.matrix(k['dataset'][0][0][0][0][0][0])
lengt = len(training_data)
```

```python
labels = k['dataset'][0][0][0][0][0][1]
```

```python
size = len(np.unique(labels))
print(np.unique(labels))
```

```python
testing_data = np.matrix(k['dataset'][0][0][1][0][0][0])
```

```python
print(testing_data)
```

```python
testing_data_labels = np.matrix(k['dataset'][0][0][1][0][0][1])
```

```python
len(testing_data_labels)
```

```python
x = tf.placeholder(tf.float32, [None, 784])


def hiddenLayer(inputT, numberOfNodes):
    '''
    Create hidden layer
    '''

    # create weights & biases
    weights = tf.Variable(tf.truncated_normal((784, numberOfNodes)), dtype = tf.float32)
    biases = tf.zeros((numberOfNodes), dtype = tf.float32)
    # output of hidden nodes
    hiddenNodes = tf.add(tf.matmul(inputT, weights), biases)
    hiddenOutput = tf.nn.sigmoid(hiddenNodes)
    
    return hiddenOutput

def outputLayer(hiddenOutput, numberOfClass):
    '''
    Create output layer
    hiddenOutput: output from hidden layer
    numOfClass: number of classes (0~9)
    '''
    inputSize = hiddenOutput.get_shape().as_list()[1]
    # create weights & biases
    weights = tf.Variable(tf.truncated_normal((inputSize, numberOfClass)), dtype = tf.float32)
    biases = tf.zeros((numberOfClass), dtype = tf.float32)
    # output
    output = tf.add(tf.matmul(hiddenOutput, weights), biases)
    
    return output

def build_nn(inputT, numberOfNodes, numberOfClass):
    '''
    build fully connected neural network
    '''
    # fully_connected layers
    fc1 = hiddenLayer(inputT, numberOfNodes)
    output = outputLayer(fc1, numberOfClass)
    
    return output

y = build_nn(x, 256, size)
y_ = tf.placeholder(tf.float32, [None, size])
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y_, logits=y))
#loss = tf.nn.l2_loss(y_ - y)
train_step = tf.train.AdamOptimizer(0.01).minimize(loss)

```

```python

sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
```

```python



A = np.hstack((training_data, labels))
#z = np.matlib.zeros((len(labels), size))
#for i in range(len(labels)):
        #z[i, labels[i, 0]] = 1

for _ in range(10000):
    print(_)
    batch_xs_ys = A[np.random.choice(A.shape[0], 1000), :]
    batch_xs = batch_xs_ys[:, :-1]
    batch_ys = batch_xs_ys[:, -1]
    z = np.matlib.zeros((len(batch_ys), size))
    for i in range(len(batch_ys)):
        z[i, batch_ys[i, 0]] = 1
    
    sess.run(train_step, feed_dict={x: batch_xs, y_: z})
    print(sess.run(loss, feed_dict={x: batch_xs, y_: z}))
```

```python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

z = np.matlib.zeros((len(testing_data_labels), size))
for i in range(len(testing_data_labels)):
    z[i, testing_data_labels[i, 0]] = 1
        
print(sess.run(accuracy, feed_dict={x: testing_data, y_: z})) #94% Accuracy with MNIST
#68% Accuracy with Letters
#57% accuracy with both

```
